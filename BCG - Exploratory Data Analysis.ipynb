```python
def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distribution in a stacked histogram of churned or retained company
    """
    
    # Create a temporal DataFrame with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    
    # X-axis label
    ax.set_xlabel(column)
    
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
     
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])
```

<h1 style="font-size: 20px;">Investigating Hypothesis</h1>


Now, let's understnd customers sensitiviy to price. When price increased, some customers left, therefore we can say these customers are sensitive to price as the increase in prices lead to customer churn. However, if there was no change or a decrease in price and customers still switch to another provider, it will be harder to say these customers are price sensitive.

So, since we have each of the companies consumption data for 2015, we can create new features to measure price sensitivty using the average of the year, the last 6 months and the last 3 months respectively.

```python
# Let's create it...
# I explicitly set 'numeric_only' parameter to 'True' in the 'mean()'
# to ensure only numeric columns are considered in mean calc...
# regardless of the default settings in future Pandas versions.

mean_year = price_df.groupby(['id']).mean(numeric_only=True).reset_index()
mean_6months = price_df[price_df['price_date'] > '2015-06-01'].groupby(['id']).mean(numeric_only=True).reset_index()
mean_3months = price_df[price_df['price_date'] > '2015-10-01'].groupby(['id']).mean(numeric_only=True).reset_index()
```

```python
mean_year.head()
```

```python
# Let's quickly rename the columns of mean - 1 year period
mean_year = mean_year.rename(

    columns={
        "price_off_peak_var": "mean_year_price_off_peak_var",
        "price_peak_var": "mean_year_price_peak_var",
        "price_mid_peak_var": "mean_year_price_mid_peak_var",
        "price_off_peak_fix": "mean_year_price_off_peak_fix",
        "price_peak_fix": "mean_year_price_peak_fix",
        "price_mid_peak_fix": "mean_year_price_mid_peak_fix"
    }
)
```

```python
mean_year["mean_year_price_off_peak"] = mean_year["mean_year_price_off_peak_var"] + mean_year["mean_year_price_off_peak_fix"]
mean_year["mean_year_price_peak"] = mean_year["mean_year_price_peak_var"] + mean_year["mean_year_price_peak_fix"]
mean_year["mean_year_price_med_peak"] = mean_year["mean_year_price_mid_peak_var"] + mean_year["mean_year_price_mid_peak_fix"]
```

```python
# Renaming the columns of mean - 6 month period
mean_6months = mean_6months.rename(

    columns={
        "price_off_peak_var": "mean_year_price_off_peak_var",
        "price_peak_var": "mean_year_price_peak_var",
        "price_mid_peak_var": "mean_year_price_mid_peak_var",
        "price_off_peak_fix": "mean_year_price_off_peak_fix",
        "price_peak_fix": "mean_year_price_peak_fix",
        "price_mid_peak_fix": "mean_year_price_mid_peak_fix"
    }
)

mean_6months["mean_year_price_off_peak"] = mean_6months["mean_year_price_off_peak_var"] + mean_6months["mean_year_price_off_peak_fix"]
mean_6months["mean_year_price_peak"] = mean_6months["mean_year_price_peak_var"] + mean_6months["mean_year_price_peak_fix"]
mean_6months["mean_year_price_med_peak"] = mean_6months["mean_year_price_mid_peak_var"] + mean_6months["mean_year_price_mid_peak_fix"]
```

```python
# Renaming the columns of mean - 3 month period
mean_3months = mean_3months.rename(

    columns={
        "price_off_peak_var": "mean_year_price_off_peak_var",
        "price_peak_var": "mean_year_price_peak_var",
        "price_mid_peak_var": "mean_year_price_mid_peak_var",
        "price_off_peak_fix": "mean_year_price_off_peak_fix",
        "price_peak_fix": "mean_year_price_peak_fix",
        "price_mid_peak_fix": "mean_year_price_mid_peak_fix"
    }
)

mean_3months["mean_year_price_off_peak"] = mean_3months["mean_year_price_off_peak_var"] + mean_3months["mean_year_price_off_peak_fix"]
mean_3months["mean_year_price_peak"] = mean_3months["mean_year_price_peak_var"] + mean_3months["mean_year_price_peak_fix"]
mean_3months["mean_year_price_med_peak"] = mean_3months["mean_year_price_mid_peak_var"] + mean_3months["mean_year_price_mid_peak_fix"]
     
```

```python
# Lastly, let's merge everything into 1 DataFrame
price_features = pd.merge(mean_year, mean_6months, on='id')
price_features = pd.merge(price_features, mean_3months, on='id')
```

```python
price_features.head()
```

```python
print("Now, we have to merge our update price feature dataset with the churn variable to check if price sensitivity actually has any correlation with churn.")


price_churn = pd.merge(price_features, client_df[['id', 'churn']], on='id')
price_churn.head()
```

```python
# Sidenote: Again, I ensured my code is future-proof by explicitly specifying
# the 'numeric_only' parameter.

corr = price_churn.corr(numeric_only=True)

# Plot correlation
plt.figure(figsize=(20, 18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, cmap='Greens', annot=True, annot_kws={'size': 10})

# Axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
print("Looking above, we see the correlation between price variabels with each other alongwith churn.\n")

print("Although, the correlation between churn and price variabels are very low - meaning customers' churn is NOT sensitive to price change.\n")

print("Moving on now. We can merge client data with price churn data for modelling.")
```

```python
churn_data = pd.merge(client_df.drop(columns=['churn']), price_churn, on='id')
```
